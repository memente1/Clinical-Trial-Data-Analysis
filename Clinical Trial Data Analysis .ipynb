#### Check for Outliers:
Outliers can be detected using statistical methods such as the Interquartile Range (IQR) or Z-score.

```python
import numpy as np

# Detect outliers using IQR
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))
    return outliers
# Detect outliers in numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
outliers = df[numeric_cols].apply(detect_outliers_iqr)
```

```python
# Remove outliers
df_no_outliers = df[~(outliers.any(axis=1))]

# Use the cleaned data for further analysis
df_final = df_no_outliers
```

## Inferential Analysis:
#### 1. Perform a paired t-test to compare the baseline and post-treatment blood pressure values to determine the efficacy of the treatment.
><p style="line-height: 1.5; font-size: 18px;"> The paired t-test will help us determine if there is a statistically significant difference between the baseline and post-treatment blood pressure values.</p>

```python
from scipy.stats import ttest_rel

# Perform paired t-test
t_stat, p_value = ttest_rel(df_final['Baseline_BP'], df_final['Post_Treatment_BP'])

print(f'Paired t-test results:')
print(f't-statistic: {t_stat}')
print(f'p-value: {p_value}')
```

><p style="line-height: 1.5; font-size: 18px;"> The t-statistic value of 35.41 is extremely high. This indicates that there is a large difference between the baseline and post-treatment blood pressure values. The higher the absolute value of the t-statistic, the greater the evidence against the null hypothesis (which in this case, the null hypothesis would be that there is no difference between baseline and post-treatment blood pressure). </p>

><p style="line-height: 1.5; font-size: 18px;"> The p-value is approximately 9.83e-94, which is an extremely small value (essentially zero for practical purposes).
this small value indicates that the probability of observing such a large difference (or larger) between the baseline and post-treatment blood pressure values, assuming the null hypothesis is true, is almost nonexistent.</p>

><p style="line-height: 1.5; font-size: 18px;"> Typically, a p-value less than 0.05 is considered statistically significant. Here, the p-value is much smaller than 0.05, providing strong evidence to reject the null hypothesis which stipulates that There is no difference in the mean blood pressure values before (baseline) and after the treatment. In other words, the null hypothesis states that the treatment has no effect on blood pressure, and any observed differences between the baseline and post-treatment blood pressure values are due to random chance.</p>

### Conclusion:
><p style="line-height: 1.5; font-size: 18px;"> - Statistical Significance: The results are statistically significant. There is a very strong evidence that the treatment has an effect on reducing blood pressure.><p 
  
><p style="line-height: 1.5; font-size: 18px;"> - Efficacy of the Treatment: Given the high t-statistic and extremely low p-value, we can conclude that the treatment is effective in significantly reducing blood pressure in patients.><p 

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table of side effects by treatment group
contingency_table = pd.crosstab(df['Side_Effects'], df['Treatment_Group'])

# Perform the Chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Output the results
print(f'Chi-square test results:')
print(f'Chi-square statistic: {chi2_stat}')
print(f'p-value: {p_val}')
print(f'Degrees of freedom: {dof}')
print('Expected frequencies:')
print(expected)
```

### Regression and Predictive Modeling:
Perform regression models to identify predictors of efficacy.
Validate model assumptions and performance.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['BP_Diff'] = df['Baseline_BP'] - df['Post_Treatment_BP']

# Convert categorical variables to numeric
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})

# Select predictors and outcome
X = df[['Age', 'Gender', 'Baseline_BP', 'Adherence', 'Follow_Up_Duration']]
y = df['BP_Diff']

# Add a constant to the model (for the intercept)
X = sm.add_constant(X)
```

### Fit the Regression Model

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Fit the regression model
model = sm.OLS(y_train, X_train).fit()

# Summarize the model
print(model.summary())
```

### Validate Model Assumptions

```python
# Linearity
plt.scatter(model.fittedvalues, y_train)
plt.xlabel('Fitted values')
plt.ylabel('Observed values')
plt.title('Linearity Check')
plt.show()

# Residuals vs Fitted values (Homoscedasticity)
plt.scatter(model.fittedvalues, model.resid)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted values')
plt.show()

# Q-Q plot (Normality)
sm.qqplot(model.resid, line='s')
plt.title('Normality of Residuals')
plt.show()

# Durbin-Watson test (Independence of residuals)
from statsmodels.stats.stattools import durbin_watson
dw_stat = durbin_watson(model.resid)
print(f'Durbin-Watson statistic: {dw_stat}')
```

### Assess Model Performance
Evaluate model performance using R-squared and RMSE

```python
# Predict on test set
y_pred = model.predict(X_test)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'RMSE: {rmse}')

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

### RMSE (Root Mean Squared Error):
><p style="line-height: 1.5; font-size: 18px;"> - Value: 6.802928263613402

><p style="line-height: 1.5; font-size: 18px;"> Interpretation: The RMSE is a measure of the differences between the predicted values by the model and the actual values. An RMSE of 6.80 indicates the average magnitude of the prediction errors in the units of the outcome variable (blood pressure difference).</p>

### R-squared:
><p style="line-height: 1.5; font-size: 18px;"> - Value: -0.021880511871517516
><p style="line-height: 1.5; font-size: 18px;"> Interpretation: R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variables. An R-squared value of -0.022 suggests that the model does not explain any of the variability in the outcome variable. In fact, a negative R-squared is a sign that the model is worse than a horizontal line (which would have an R-squared of 0).</p>

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF for each predictor
vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

### The Variance Inflation Factor (VIF) results show that multicollinearity is not a significant issue in your model:

><p style="line-height: 1.5; font-size: 18px;"> - const: The intercept term has a very high VIF, which is expected and typically ignored.
><p style="line-height: 1.5; font-size: 18px;"> - Other Variables: All the predictors (Age, Gender, Baseline_BP, Adherence, Follow_Up_Duration) have ><p style="line-height: 1.5; font-size: 18px;"> VIF values close to 1, indicating that multicollinearity is not a concern.
Since multicollinearity is not an issue, let's proceed with fitting a different model to see if it improves the prediction performance.

#### Fit a Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

# Prepare the data
X = df[['Age', 'Gender', 'Baseline_BP', 'Adherence', 'Follow_Up_Duration']]
y = df['BP_Diff']

# Add a constant to the model (for the intercept)
X = sm.add_constant(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize the Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)

# Fit the model on the training data
rf_model.fit(X_train, y_train)

# Predict on the test set
y_pred_rf = rf_model.predict(X_test)

# Calculate RMSE
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print(f'Random Forest RMSE: {rmse_rf}')

# Calculate R-squared
r2_rf = r2_score(y_test, y_pred_rf)
print(f'Random Forest R-squared: {r2_rf}')

# Cross-validation
cv_scores = cross_val_score(rf_model, X, y, cv=5, scoring='neg_mean_squared_error')
cv_rmse = np.sqrt(-cv_scores)
print(f'Cross-validated RMSE: {cv_rmse.mean()}')
```

><p style="line-height: 1.5; font-size: 18px;"> - Cross-validated RMSE: A cross-validated RMSE of 7.48 suggests that the model's performance is consistent across different subsets of the data but still indicates higher prediction errors.</p>

#### Expected frequencies:
><p style="line-height: 1.5; font-size: 18px;"> The expected frequencies are the counts we would expect in each category if there were no association between treatment group and side effects. These are calculated based on the marginal totals of the contingency table.</p>

### Conclusion:
><p style="line-height: 1.5; font-size: 18px;"> - Statistical Significance: The p-value (0.6423874114327723) is much larger than the common significance level (0.05). Therefore, we fail to reject the null hypothesis.
><p style="line-height: 1.5; font-size: 18px;"> - Null Hypothesis: The null hypothesis for the chi-square test is that there is no association between the treatment group and side effects.</p>
><p style="line-height: 1.5; font-size: 18px;"> - Interpretation: The results indicate that there is no statistically significant difference in the distribution of side effects between the treatment and placebo groups. The observed differences in side effects are likely due to random chance rather than the effect of the treatment.</p>
